```r
data_breast_csv <- read.csv2("breast-cancer-wisconsin.csv",  header = TRUE, dec= ".", stringsAsFactors=FALSE, sep = ",")
```

```r
tail(data_breast_csv)
```

```r
set.seed(1000)
```

```r
head(data_breast_csv[,1:30])
```

```r
data_breast <- unique(data_breast_csv[,1:30])
breast_matrix <- as.matrix(data_breast)
```

```r
library(ggplot2)
library(Rtsne)
```

```r
tsne_out <- Rtsne(breast_matrix, perplexity = 30, theta = 0.0, max_iter = 1000) # Run TSNE
```

```r
qplot(tsne_out$Y[,1], tsne_out$Y[,2], colour = data_breast_csv$label)
```

```r
tsne_out_10 <- Rtsne(breast_matrix, perplexity = 10, theta = 0.0, max_iter = 1000) # Run TSNE
```

```r
qplot(tsne_out_10$Y[,1], tsne_out_10$Y[,2], colour = data_breast_csv$label)
```

```r
tsne_out_50 <- Rtsne(breast_matrix, perplexity = 50, theta = 0.0, max_iter = 1000) # Run TSNE
```

```r
qplot(tsne_out_50$Y[,1], tsne_out_50$Y[,2], colour = data_breast_csv$label)
```

```r
tsne_out_10_3D <- Rtsne(breast_matrix, perplexity = 10, theta = 0.0, max_iter = 1000, dims =3) # Run TSNE
```

```r
library(rgl)
plot3d(tsne_out_50$Y[,1], tsne_out_50$Y[,2], tsne_out_50$Y[,3], colour = data_breast_csv$label, type="s", size=0.75, lit=FALSE)
```
