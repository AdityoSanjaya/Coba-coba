```r
data_wine_csv <- read.csv2("winequality-small.csv",  header = TRUE,  dec= ".", stringsAsFactors=FALSE, sep = ",")
```

```r
tail(data_wine_csv)
```

```r
set.seed(1000)
```

```r
data_wine <- unique(data_wine_csv[,1:10])
wine_matrix <- as.matrix(data_wine)
```

```r
library(ggplot2)
library(Rtsne)
```

```r
tsne_out <- Rtsne(wine_matrix, perplexity = 10, theta = 0.0, max_iter = 10000) # Run TSNE
```

```r
qplot(tsne_out$Y[,1], tsne_out$Y[,2] ,
      colors = data_wine_csv$quality,
      type="s",
      size=0.75,
      lit=FALSE)
```
