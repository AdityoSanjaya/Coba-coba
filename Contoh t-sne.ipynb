```r
library(Rtsne)
set.seed(123)
```

```r
iris_unique <- unique(iris[,1:4])
iris_matrix <- as.matrix(iris_unique)
```

```r
tsne_out <- Rtsne(iris_matrix, perplexity = 30, theta = 0.0, max_iter = 10000) # Run TSNE
```

```r
plot(tsne_out$Y,col = iris$Species)
```
